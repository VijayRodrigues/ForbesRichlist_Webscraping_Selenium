```python
import selenium
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
```

```python
driver = webdriver.Chrome(r"C:\Users\HP\Downloads\chromedriver_win32\chromedriver.exe")

url = "https://www.forbes.com/?sh=1b888df32254"
driver.get(url)
driver.maximize_window()
```

```python
drop_hover = driver.find_element_by_xpath("//button[@class='icon--hamburger']").click()
```

```python

#object of ActionChains
a = ActionChains(driver)
#identify element
m = driver.find_element_by_link_text("Billionaires")
#hover over element
a.move_to_element(m).perform()
#identify sub menu element
n = driver.find_element_by_link_text("World's Billionaires")
# hover over element and click
a.move_to_element(n).click().perform()
print("Page title: " + driver.title)
#close browser
#driver.close()
```

```python
stop_hover = driver.find_element_by_xpath("//div[@class='personName']").click()
```

```python
begin = time.time()

top_ranks=[]
top_names=[]
net_worth=[]
total_age=[]
citizenship=[]
source=[]
industry=[]

for page in range(1,7):
    top_rank = driver.find_elements_by_xpath("//div[@class='rank']")
    for rank in top_rank:
        top_ranks.append(rank.text)

    top_name = driver.find_elements_by_xpath("//div[@class='personName']")
    for name in top_name:
        top_names.append(name.text)

    top_net = driver.find_elements_by_xpath("//div[@class='netWorth']")
    for net in top_net:
        net_worth.append(net.text)

    top_age = driver.find_elements_by_xpath("//div[@class='age']")
    for age in top_age:
        total_age.append(age.text)

    top_country = driver.find_elements_by_xpath("//div[@class='countryOfCitizenship']")
    for cntry in top_country:
        citizenship.append(cntry.text)    

    top_source = driver.find_elements_by_xpath("//div[@class='source']")
    for srce in top_source:
        source.append(srce.text)

    top_industry = driver.find_elements_by_xpath("//div[@class='category']")
    for indus in top_industry:
        industry.append(indus.text)  
    
next_button = driver.find_element_by_xpath("//button[@class='pagination-btn pagination-btn--next ']").click()

end = time.time()

print(f"Total runtime of the program is {end - begin}")
```

```python
print(len(top_ranks))
print(len(top_names))
print(len(net_worth))
print(len(total_age))
print(len(citizenship))
print(len(source))
print(len(industry))
```

```python
all_details = {
    "Total Rank":top_ranks,
    "Name of the Billionaire":top_names,
    "Net Worth":net_worth,
    "Age":total_age,
    "Citizenship":citizenship,
    "Source of Income":source,
    "Industry":industry
}
```

```python
pd.set_option("display.max_rows",None)
billionaire_details = pd.DataFrame(all_details)
billionaire_details.sort_values(by=["Total Rank"], ascending=True).drop_duplicates()
```
